Функция $f(x) = C^T \cdot X + \cfrac{1}{3}||X||^{3}_{2}$

Аналитически нашли, что min функции достигается при:<br><br> $$\overrightarrow{X}\cdot ||\overrightarrow{X}||_2 + \overrightarrow{C} = 0$$<br>

Пусть $\overrightarrow{C} = \begin{pmatrix} a \\ b \end{pmatrix}, \quad \overrightarrow{X} = \begin{pmatrix} x1 \\ x2 \end{pmatrix}$, тогда перепишем условие в виде:<br>
$$\begin{equation}
\left\{ \begin{aligned}
   a + x_1 \cdot \sqrt{x_1^2+x_2^2} &=0 \\
   b + x_2 \cdot \sqrt{x_1^2+x_2^2} &=0
\end{aligned} \right.
\end{equation}
$$
<br>
Отсюда выразим значения $x1$ и $x2$:
$$\begin{equation}
\left\{ \begin{aligned}
   x_1 &= \cfrac{-a}{\sqrt[4]{a^2+b^2}} \\
   x_2 &= \cfrac{-b}{\sqrt[4]{a^2+b^2}}
\end{aligned} \right.
\end{equation}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Определение функции
def f(x, y, a, b):
    return a*x + b*y + (1/3) * (x**2 + y**2)**1.5

# Параметры функции
a = 16
b = 12

# Аналитическое решение
X_min_teor = -a / (a**2 + b**2)**(1/4)
Y_min_teor = -b / (a**2 + b**2)**(1/4)

# Создание сетки значений
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)

# Вычисление значений функции на сетке
Z = f(X, Y, a, b)

# Построение линий уровня
plt.contour(X, Y, Z, levels=20)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Линии уровня функции')
plt.colorbar()

# Отображение аналитического минимума на графике
plt.plot(X_min_teor, Y_min_teor, 'ro', label='Аналитический минимум')
plt.legend()

# Численное нахождение минимума на сетке
Z_min_index = np.unravel_index(np.argmin(Z, axis=None), Z.shape)
X_min_graphic = X[Z_min_index]
Y_min_graphic = Y[Z_min_index]

# Отображение численного минимума на графике
plt.plot(X_min_graphic, Y_min_graphic, 'bx', label='Численный минимум')
plt.legend()

# Вывод координат
print(f'Координаты точки минимума аналитические: X_min = {X_min_teor:.3f}, Y_min = {Y_min_teor:.3f}')
print(f'Координаты точки минимума на графике: X_min = {X_min_graphic:.3f}, Y_min = {Y_min_graphic:.3f}')

plt.show()
```
